Finding Missing Values

```python
import pandas as pd
data={
    "Name":["Shree","Anu",None,"Ritu","Gyan"],
    "City":["Indore",None,"Dhar",None,"Delhi"],
    "DOB":[2003,None,1998,1995,2000],
    "Salary":[20000,10000,30000,None,23000]
}
df=pd.DataFrame(data)
print(df.isnull())         #true means none value
print()
print("No. of none values")
print(df.isnull().sum())
```

Handling Missing Values

```python
import pandas as pd
data={
    "Name":["Shree","Anu",None,"Ritu","Gyan"],
    "City":["Indore",None,None,"Dhar","Delhi"],
    "DOB":[2003,None,None,1995,2000],
    "Salary":[20000,10000,None,None,23000]
}
df=pd.DataFrame(data)

#df.dropna(axis=0,inplace=True)     #remove each row which has none value
print(df) 
print()
#print(df.dropna(how='all'))    #remove row which has all elements null
print()
print(df.dropna(thresh=2))    #remove row which has 2 null elements 
print()
```

fillna method

```python
print(df)
print()

#print(df.fillna(0,inplace=True))

'''df['DOB'].fillna(df['DOB'].mean())
df['Salary'].fillna(df['Salary'].mean())'''
g=df.fillna({
    'DOB' : '2000',
    'Salary' : df.Salary.mean(),
    'City' : 'Dhar',
    'Name' : 'Shree'
})
print(g)
print("forward fill")
#forward fill
print(df.ffill(limit=2))
print("backward fill")
#backward fill
print(df.bfill(axis='columns'))
```
