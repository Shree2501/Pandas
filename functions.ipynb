Converting to CSV file using to_csv function

```python
import pandas as pd
data={
    "Name":["Shree","Anu","Sam","Ritu","Gyan"],
    "City":["Indore","Dhar","Dhar","Dhar","Delhi"],
    "DOB":[2003,1990,1998,1995,2000],
    "Salary":[20000,10000,30000,40500,23000],
    "Currency":['INR','$','USD','$','Dollar']
}
rd=pd.DataFrame(data)
print(rd)
rd.to_csv("output.csv",index=False)    #csv file
rd.to_json("output.xlsx",index=False)    #excel file                                                                                                                                                                                                  
```

max / min function

```python
import pandas as pd
data={
    "Name":["Shree","Anu","Sam","Ritu","Gyan"],
    "Age":[10,20,30,40,20],
    "City":["Indore","Dhar","Dhar","Dhar","Delhi"]
}
df=pd.DataFrame(data)
print(df.Age.max())
print(df.min())
print(df[df.Age == df.Age.max()])
print(df[(df.Age == df.Age.max()) | (df.Age == df.Age.min())])
```

apply function

```python
import pandas as pd
data={
    "Name":["Shree","Anu","Sam","Ritu","Gyan"],
    "City":["Indore","Dhar","Dhar","Dhar","Delhi"],
    "DOB":[2003,1990,1998,1995,2000],
    "Salary":[20000,10000,30000,40500,23000]
}
df=pd.DataFrame(data)
df["Age"]=df['DOB'].apply(lambda x : 2025-x)
print(df)
print()
df["increment"]=df['Salary'].apply(lambda x: x+x/10)
print(df)
print()
df["Bonus"]=df.apply(lambda x : x['increment'] - x['Salary'] ,axis=1)
print(df)
```

Concatenation

```python
import pandas as pd
df_customer=pd.DataFrame ({
    "cust_id":[1,2,3],
    "name":["dev","raj","indra"]
})
df_order=pd.DataFrame ({
    "cust_id":[1,2,4],
    "amount":[222,333,444]
})
print(pd.concat([df_customer,df_order],ignore_index=True))
print()
print(pd.concat([df_customer,df_order],axis=1,ignore_index=True))
```
