single and multiple conditions

```python
import pandas as pd
data={
    "Name":["Shree","Anu","Sam","Ritu","Gyan"],
    "Age":[10,20,30,40,50],
    "City":["Indore","Dhar","Bihar","Patna","Delhi"]
}
df=pd.DataFrame(data)
print(df[df["Age"]>30])
print(df[(df["Age"]>10) & (df["Age"]<30)])
print(df[(df["Age"]>30) | (df["Name"]=="Anu")])
```
