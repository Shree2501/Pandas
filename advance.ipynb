```python
l=[1,2,3,4,5,6]
print(l)
def fxn(n):
    if(n%2==0):
        return True
    return False
list(filter(fxn,l))
```

```python
from functools import reduce
l=[1,2,3,4,5,6]
print(l)
def sum(x,y):
    return x+y
print([reduce(sum,l)])
```

```python
l=[2,5,24,6,20,50,35,33,22,54,7,8]
def div(n):
    if(n%5==0):
        return True
    return False
print(list(filter(div,l)))
```

```python
l=[2,5,24,6,20,50,35,33,22,54,7,8,99]
def max(x,y):
    if(x>y):
        return x
    return y
print([reduce(max,l)])
```
