```python
import pandas as pd
df=pd.read_excel("CustomerCallList.xlsx")
print(df.tail(10))
```

```python
df["Last_Name"].ffill(inplace=True)
df["Last_Name"].head(10)
```

```python
import numpy as np
df["Phone_Number"].replace("N/a" ,np.nan, inplace=True)
df['Phone_Number'].fillna("000-000-0000", inplace=True)
df['Phone_Number'].head(10)
```

```python
df['Last_Name'] = df['Last_Name'].str.replace(r'[^A-Za-z ]', '', regex=True).str.strip()
df["Last_Name"].head(10)
```

```python
df["Address"].replace("N/a" ,np.nan, inplace=True)
df['Address'].fillna("Not Available", inplace=True)
df["Address"].head(10)
```

```python
df["Paying Customer"].replace("Y","Yes",inplace=True)    
df["Paying Customer"].replace("N","No",inplace=True)
df["Paying Customer"].replace("N/a",np.nan,inplace=True)    
df["Paying Customer"].fillna("No",inplace=True)
df["Paying Customer"].head(10)
```

```python
df["Do_Not_Contact"].replace("Y","Yes",inplace=True)
df["Do_Not_Contact"].replace("N","No",inplace=True)
df["Do_Not_Contact"].fillna("No",inplace=True)
df["Do_Not_Contact"].head(10)   
```

```python
df["Not_Useful_Column"].drop(columns=["Not_Useful_Column"],inplace=True)
```
