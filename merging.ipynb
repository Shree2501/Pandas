Merging and joining

```python
import pandas as pd
df_customer=pd.DataFrame ({
    "cust_id":[1,2,3],
    "name":["dev","raj","indra"]
})
df_order=pd.DataFrame ({
    "cust_id":[1,2,4],
    "amount":[222,333,444]
})
#inner join
merged=pd.merge(df_customer,df_order,on="cust_id",how="inner")
print(merged)
print()
#outer join 
merged=pd.merge(df_customer,df_order,on="cust_id",how="outer")
print(merged)
print()
#inner join
merged=pd.merge(df_customer,df_order,on="cust_id",how="left")
print(merged)
print()
#inner join
merged=pd.merge(df_customer,df_order,on="cust_id",how="right")
print(merged)
print()
```
